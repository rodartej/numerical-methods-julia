```julia
using LinearAlgebra

"""
    chebyshev_polinomio(n, x)

Evalúa el n-ésimo Polinomio de Chebyshev de primera especie T_n(x) en el punto x.
Usa la definición trigonométrica para x en [-1, 1] y la recurrencia para fuera de [-1, 1].
"""
function chebyshev_polinomio(n::Int, x::Real)

    # si n es grande (aunque la recurrencia es más estable). Usamos la recurrencia aquí.
    if n == 0
        return 1.0
    elseif n == 1
        return x
    else
        T_prev2 = 1.0 # T_0
        T_prev1 = x   # T_1
        for k = 2:n
            # relación de recurrencia
            T_k = 2 * x * T_prev1 - T_prev2
            T_prev2 = T_prev1
            T_prev1 = T_k
        end
        return T_prev1
    end
end

"""
    chebyshev_monico_polinomio(n, x)

devuelve el n-ésimo Polinomio Mónico de Chebyshev \\tilde{T}_n(x) en el punto x.
"""
function chebyshev_monico_polinomio(n::Int, x::Real)
    T_n = chebyshev_polinomio(n, x)
    if n <= 1
        return T_n # T_0 = 1, T_1(x) = x ya son mónicos
    else
        return T_n / (2^(n - 1))
    end
end

"""
    aproximacion_chebyshev(f, grado)

Calcula los coeficientes de Chebyshev para la función f(x) en [-1, 1] 
usando la integración discreta (puntos de Chebyshev-Gauss-Lobatto).

# devuelve
- Un vector de coeficientes Chebyshev (c_0, c_1, ..., c_grado).
"""
function aproximacion_chebyshev(f::Function, grado::Int)
    N = grado
    
    # Puntos de Chebyshev-Gauss-Lobatto (extremos incluidos)
    # x_j = cos(pi * j / N)
    X_k = [cos(pi * k / N) for k = 0:N]

    # Valores de la función en los puntos X_k
    F_k = f.(X_k)
    
    coeficientes = zeros(N + 1)

    for k = 0:N
        c_k = 0.0
        
        # Sumatoria discreta: Suma de la función y el polinomio en los puntos de Chebyshev
        for j = 0:N
            
            # El término 1/2 en los extremos está implícito en el método
            w_j = 1.0
            if j == 0 || j == N
                w_j = 0.5 # Los puntos extremos se ponderan a la mitad
            end
            
            # T_k(x_j) se puede calcular como cos(k * arccos(x_j)) = cos(k * pi * j / N)
            T_k_xj = cos(k * pi * j / N)
            
            c_k += w_j * F_k[j+1] * T_k_xj
        end
        
        # Factor de normalización 2/N
        coeficientes[k+1] = c_k * (2.0 / N)
    end

    return coeficientes
end
```

```julia
# ejemplo
f_ejemplo(x) = abs(x)
grado_chebyshev = 5

coefs_chebyshev = aproximacion_chebyshev(f_ejemplo, grado_chebyshev)

println("--- Aproximación Polinómica de Chebyshev (Grado $grado_chebyshev) ---")
println("Coeficientes Chebyshev (c_0, c_1, ..., c_5):")
println(coefs_chebyshev)

# se muestra el polinomio mónico
x_val = 0.6
n_monico = 3
T3_monico = chebyshev_monico_polinomio(n_monico, x_val)

println("\nPolinomio Mónico:")
println(T3_monico)
```
