### Investigación
BLAS y LAPACK son bibliotecas, de software libre, fundamentales para operaciones comunes de álgebra lineal.

Son utilizadas para resolver sistemas de ecuaciones lineales, multiplicar matrices, realizar descomposiciones (LU, QR, Cholesky, SVD) y calcular eigenvalores y eigenvectores.

#### BLAS
BLAS se centra en rutinas de bajo nivel (operaciones kernel o núcleo) que sirven para operar. Estas rutinas son el éstandar de bajo nivel para las bibliotecas de álgebra lineal. Algunas de las operaciones de BLAS, son las siguientes:
- Suma de vectores
- Multiplicación escalar
- Productos punto
- Combinaciones lineales
- Multiplicación de matrices

Su funcionalidad está organizada en tres niveles, estos niveles son con base en la complejidad de los algoritmos.
| Nivel | Tipo de operación      | Complejidad |
|:------|:-----------------------|:-------------|
| 1     | Operaciones vector–vector | $O(n)$ |
| 2     | Operaciones matriz–vector | $O(n^2)$ |
| 3     | Operaciones matriz–matriz | $O(n^3)$ |

#### LAPACK
LAPACK aborda problemas más sofisticados en el álgebra lineal numérica y usa rutinas optimizadas de BLAS. Estas rutinas son de las más importantes:
- Factorizaciones de Matrices
- Resolución directa de Sistemas Lineales $Ax = b$
- Eigenvalores y eigenvectores

#### Julia y su relación con LAPACK y BLAS
El lenguaje Julia está diseñado para interactuar directamente y aprovechar estas bibliotecas altamente optimizadas.

Julia integra de forma directa las bibliotecas BLAS y LAPACK dentro de su módulo LinearAlgebra, lo que le permite ejecutar operaciones numéricas con la misma eficiencia que lenguajes compilados como C o Fortran.

Los wrappers son interfaces que sirven de intermediarios para usar una función/biblioteca escrita en otro lenguaje y llamar una subrutina de bajo nivel. Funciones como dot, gemv! o gemm! (de BLAS) y rutinas de factorización como getrf! o geqrf! (de LAPACK) son wrappers que usa Julia para seleccionar el algoritmo según el problema. Esta arquitectura convierte a su biblioteca LinearAlgebra en una interfaz de alto nivel que conserva la velocidad y precisión de las implementaciones optimizadas de BLAS y LAPACK. El wrapper traduce los argumentos de Julia (matrices, tipos, punteros) al formato que espera la rutina en C/Fortran y devuelve el resultado listo para usarse en Julia.

> Los métodos iterativos dejan de tener sentido al perder la noción de convergencia, esto sucede porque no existe una estructura topológica que permita definir continuidad, límites o distancias entre elementos (por tanto no puedes definir una norma ni un error residual que tienda a cero) y hace que las iteraciones solo puedan oscilar o repetir valores dentro de un conjunto discreto.

Por otro lado, existe teoría de análisis numérico que sujeta este argumento de manera formal. Se requiere de forma explícita que el espacio sea un espacio normado para que los métodos iterativos tengan sentido.

Ref: Tsynkov, S. S. (2018). Numerical methods for partial differential equations (Section 6.1, p. 175). North Carolina State University. Recuperado de https://stsynkov.math.ncsu.edu/book_sample_material/Section_6.1.pdf

## 2.2 Dar ejemplos para $F = \mathbb{Z}_p$ con al menos 1000 primos distintos en sistemas de $n \times n$ con $n = 2, 3, \ldots, 100$.

### Solución usando factorización LU optimizada para números primos.
> Primero cargamos las librerías.

```julia
using LinearAlgebra
using Random
using Primes
```

> Creamos una función para generar primos, usa nextprime que es de la librería de Primes.

```julia
function generar_primos(count::Int)
    ps = Vector{Int}(undef, count)
    p = nextprime(2) # nextprime es una función de la librería Primes para obtener el siguiente primo
    for i in 1:count
        ps[i] = p
        p = nextprime(p+1)
    end
    return ps
end
```

> La operación módulo $p$ toma un número y devuelve el resto de dividirlo entre $p$, manteniendo así todos los valores dentro del conjunto $\{0, 1, 2, \dots, p - 1\}$.
Aseguramos con esta función que todos sus valores queden dentro del rango válido del campo finito $\mathbb{Z}_p$.

```julia
function modp!(X::AbstractArray{Int}, p::Int)
    X .%= p
    @inbounds @simd for i in eachindex(X) # inbounds y simd sólo son para optimización
        if X[i] < 0; X[i] += p; end
    end
    return X
end
```

> Esto es el inverso multiplicativo modular.

> Ref: invmod » Julia Functions. https://www.jlhub.com/julia/manual/en/function/invmod. Accedido 6 de octubre de 2025.


```julia
@inline invmod_p(a::Int, p::Int) = invmod(mod(a, p), p)   # inverso x: a*x ≡ 1 (mod p)
```

> Aplicamos la factorización LU, a diferencia de LU normal, aquí todas las operaciones se realizan módulo $p$, lo que significa que cada suma, resta, multiplicación o división se reduce al resto de dividir entre $p$. La “división” se sustituye por el inverso modular, ¿por qué? Porque así garantizamos que cada elemento tenga un equivalente multiplicativo válido dentro del campo.

> Después de esta función, descomponemos la matriz A en tres componentes:
- $L$: matriz triangular inferior con unos en la diagonal,

- $U$: matriz triangular superior,

- $P$: es vector de que registra los intercambios de filas necesarios para mantener la matriz invertible mod $p$.

```julia
function lup_modp(A::Matrix{Int}, p::Int)
    n = size(A,1); @assert size(A,2)==n
    A = copy(A); modp!(A,p)
    L = Matrix{Int}(I, n, n)   # diag(L)=1
    U = zeros(Int, n, n)
    P = collect(1:n)

    for k in 1:n
        # pivote != 0 en col k
        piv = 0
        for r in k:n
            if A[r,k] % p != 0
                piv = r; break
            end
        end
        piv == 0 && error("Matriz singular mod $p (columna $k)")

        # swap filas (A y L parcial) + permutación
        if piv != k
            A[k,:], A[piv,:] = A[piv,:], A[k,:]
            L[k,1:k-1], L[piv,1:k-1] = L[piv,1:k-1], L[k,1:k-1]
            P[k], P[piv] = P[piv], P[k]
        end

        # fila k de U
        for j in k:n
            U[k,j] = (A[k,j] - sum(L[k,1:k-1] .* U[1:k-1,j])) % p
        end
        invpiv = invmod_p(U[k,k], p)

        # columna k de L (debajo de diag)
        for i in k+1:n
            L[i,k] = ((A[i,k] - sum(L[i,1:k-1] .* U[1:k-1,k])) * invpiv) % p
        end
    end
    modp!(L,p); modp!(U,p)
    return L, U, P
end
```

> Hacemos la sustitución hacia adelante en el campo finito $\mathbb{Z}_p$.
Usa la matriz $L$ (triangular inferior con unos en la diagonal) y el vector $Pb$ (resultado de aplicar la permutación $P$ a $b$) para calcular el vector $y$ en el sistema $L y \equiv P b \pmod{p}$.
En cada paso, el valor de $y_i$ se obtiene restando la suma de los productos anteriores y aplicando la operación módulo $p$, para que todos los cálculos permanezcan dentro del campo finito.

```julia
function forward_substitution_modp(L::Matrix{Int}, Pb::Vector{Int}, p::Int)
    n = length(Pb); y = zeros(Int,n)
    for i in 1:n
        s = 0
        @inbounds for j in 1:i-1
            s += L[i,j]*y[j]
        end
        y[i] = (Pb[i] - s) % p
    end
    return y
end
```

> Hacemos la sustitución hacia atrás.

```julia
function backward_substitution_modp(U::Matrix{Int}, y::Vector{Int}, p::Int)
    n = length(y); x = zeros(Int,n)
    for i in n:-1:1
        s = 0
        @inbounds for j in i+1:n
            s += U[i,j]*x[j]
        end
        x[i] = ((y[i] - s) * invmod_p(U[i,i], p)) % p
    end
    return x
end
```

> Toma todo el proceso (factorizar, reordenar, sustituir hacia adelante y hacia atrás) y lo agrupa en una sola operación.

```julia
function solve_lu_modp(A::Matrix{Int}, b::Vector{Int}, p::Int)
    n = size(A,1)
    @assert size(A,2)==n && length(b)==n "Dimensiones inconsistentes"
    L,U,P = lup_modp(A,p)
    Pb = b[P]; modp!(Pb,p)
    y = forward_substitution_modp(L, Pb, p)
    x = backward_substitution_modp(U, y, p)
    return x, L, U, P
end
```

> Generamos 1000 primos.

> Seleccionamos las dimensiones de la matriz.

```julia
println("Generando los primeros 1000 números primos...")
primeros_1000_primos = generar_primos(1000)
println("Listo.\n")

dimensiones_a_probar = [3, 10, 50, 100]
primos_a_probar = [5, 101, primeros_1000_primos[500], primeros_1000_primos[end]]

rng = MersenneTwister(1234) # semilla que viene de Random para producción reproducible de números aleatorios
```

> Probamos distintas dimensiones y campos finitos ℤₚ.  

> En cada caso se crea un sistema con una solución conocida para verificar el método.  

> Si la matriz A no se puede invertir (caso singular), se vuelve a intentar hasta 20 veces.  

> Después se revisa que la factorización y la solución sean correctas.

```julia
for n in dimensiones_a_probar
    for p in primos_a_probar
        println("\n", "-"^60)
        println("Ejemplo: n = $n   |   p = $p   (campo ℤ_$p)")
        println("-"^60)

        # Generamos un sistema con solución conocida x_true y construimos b = A*x_true (mod p).
        # Si A sale singular, reintentamos algunas veces.
        attempts = 0
        max_attempts = 20
        solved = false
        while attempts < max_attempts && !solved
            attempts += 1
            A = rand(rng, 0:p-1, n, n)
            x_true = rand(rng, 0:p-1, n)
            b = mod.(A * x_true, p)
            try
                x, L, U, P = solve_lu_modp(A, b, p)
                # verificaciones:
                ok_factor = ((L*U) .% p) == (A[P,:] .% p)
                ok_system = ((A * x) .% p) == (b .% p)
                if ok_factor && ok_system
                    println("✅ Correcto en intento $attempts")
                    println("x verdadera:   ", x_true)
                    println("x calculada:   ", x)
                    solved = true
                end
            catch
                # singular; reintentar
            end
        end
        if !solved
            println("⚠️  No se logró un sistema invertible tras $max_attempts intentos (p=$p, n=$n).")
        end
    end
end
```

> Aunque los resultados parecen distintos, son correctos porque en aritmética modular los números que difieren en un múltiplo de $p$ representan el mismo valor dentro del campo finito $\mathbb{Z}_p$.

## Descomposición de valores singulares (SVD)

> Creamos la función PCA para nuestra matriz. "Principal Component Analysis (PCA) busca los ejes principales de la variabilidad en un conjunto de datos."

> La importancia de esta función radica en obtener la dirección de variabilidad significativa.

### Funciones.

```julia
function pca_svd_imagen(A::AbstractMatrix{<:Real}, k::Integer)
    F = svd(A)
    U, S, Vt = F.U[:, 1:k], F.S[1:k], F.Vt[1:k, :]
    return U * Diagonal(S) * Vt
end
```

> Puse comentarios para cada etapa que está a continuación:

1. Primero cargamos la imagen, luego la separamos para los canales de color.

2. Usamos SVD para cada canal.

3. Definir cuántas componentes (k) se usarán para reconstruir la imagen con diferentes niveles de compresión.

4. Mostrar imagen original

5. Crear lista plots para después mostrar los plots juntos

6. Reconstruir la imagen con k componentes

7. Mostrar plot por cada k

8. Mostrar todos

```julia
function comprimir_imagen_pca_svd(ruta_imagen::String)
    println("Cargando imagen desde: ", ruta_imagen)
    try
        # 1. cargar imagen
        img = load(ruta_imagen)
        println("Imagen cargada con tamaño ", size(img))

        # 2. separar canales
        R = Float32.(red.(img))
        G = Float32.(green.(img))
        B = Float32.(blue.(img))

        # 3. componentes a usar
        base_dim = min(size(R,1), size(R,2))
        max_k = min(base_dim, 512)
        ks = collect(potencias_de_dos(max_k))
        if isempty(ks)
            ks = [1]
        end
        println("Componentes a probar: ", ks)

        # 4. imagen original
        p_original = plot(img, title="Original", aspect_ratio=:equal, axis=nothing)
        display(p_original)
        sleep(1)

        # 5. lista de plots
        plots_list = [p_original]

        # 6. reconstrucción de la imagen con diferentes k
        for k in ks
            kk = min(k, base_dim)

            r_rec = pca_svd_imagen(R, kk)
            g_rec = pca_svd_imagen(G, kk)
            b_rec = pca_svd_imagen(B, kk)

            img_rec = colorview(RGB,
                                clamp.(r_rec, 0f0, 1f0),
                                clamp.(g_rec, 0f0, 1f0),
                                clamp.(b_rec, 0f0, 1f0))

            # 7. mostrar plot por cada k
            p = plot(img_rec, title="k = $kk", aspect_ratio=:equal, axis=nothing)
            display(p)

            push!(plots_list, p)

            sleep(1)
        end

        # 8. mostrar todos
        n_plots = length(plots_list)
        n_cols = ceil(Int, sqrt(n_plots))
        n_rows = ceil(Int, n_plots / n_cols)

        display(plot(plots_list..., layout=(n_rows, n_cols), size=(1400, 900)))


    catch e
        if e isa SystemError
            println("\nNo se pudo encontrar la imagen en la ruta especificada.")
        else
            rethrow(e)
        end
    end
end
```

## Matriz de correlación

> La idea es la misma, solo cambia cómo obtenemos las direcciones principales (los componentes).

> Hacemos la PCA usando la matriz de correlación.

### Funciones

```julia
function pca_corr_imagen(A::AbstractMatrix{<:Real}, k::Integer)
    # 1. centrar matriz
    A_centrada = A .- mean(A, dims=1)
    
    # 2. calcular matriz de correlación
    C = cor(A_centrada)
    
    # 3. eigenvalores y eigenvectores
    vals, vecs = eigen(C)
    
    # 4. ordenamos de mayor a menor
    orden = sortperm(vals, rev=true)
    vals = vals[orden]
    vecs = vecs[:, orden]
    
    # 5. seleccionamos k componentes principales
    k = min(k, size(vecs, 2))
    P = vecs[:, 1:k]
    
    # 6. proyectamos y reconstruimos
    proyeccion = A_centrada * P
    A_rec = proyeccion * P' .+ mean(A, dims=1)
    
    return A_rec
end
```

> Reconstrucción de la imagen.

```julia
function comprimir_imagen_pca_corr(ruta_imagen::String)
    println("Cargando imagen desde: ", ruta_imagen)
    try
        img = load(ruta_imagen)
        println("Imagen cargada con tamaño ", size(img))

        R = Float32.(red.(img))
        G = Float32.(green.(img))
        B = Float32.(blue.(img))

        base_dim = min(size(R,1), size(R,2))
        max_k = min(base_dim, 512)
        ks = collect(potencias_de_dos(max_k))
        if isempty(ks)
            ks = [1]
        end
        println("Componentes a probar: ", ks)

        p_original = plot(img, title="Original", aspect_ratio=:equal, axis=nothing)
        display(p_original)
        sleep(1)

        plots_list = [p_original]

        for k in ks
            kk = min(k, base_dim)

            r_rec = pca_corr_imagen(R, kk)
            g_rec = pca_corr_imagen(G, kk)
            b_rec = pca_corr_imagen(B, kk)

            img_rec = colorview(RGB,
                                clamp.(r_rec, 0f0, 1f0),
                                clamp.(g_rec, 0f0, 1f0),
                                clamp.(b_rec, 0f0, 1f0))

            p = plot(img_rec, title="k = $kk", aspect_ratio=:equal, axis=nothing)
            display(p)
            push!(plots_list, p)

            sleep(1)
        end

        n_plots = length(plots_list)
        n_cols = ceil(Int, sqrt(n_plots))
        n_rows = ceil(Int, n_plots / n_cols)

        display(plot(plots_list..., layout=(n_rows, n_cols), size=(1400, 900)))

    catch e
        if e isa SystemError
            println("\nNo se pudo encontrar la imagen en la ruta especificada.")
        else
            rethrow(e)
        end
    end
end
```

## Matriz de covarianza

> Hacemos la PCA usando la matriz de covarianza.

### Funciones

```julia
function pca_cov_imagen(A::AbstractMatrix{<:Real}, k::Integer)
    # 1. centrar la matriz
    A_centrada = A .- mean(A, dims=1)
    
    # 2. matriz de covarianza
    C = cov(A_centrada)
    
    # 3. eigenvalores y eigenvectores
    vals, vecs = eigen(C)
    
    # 4. ordenar de mayor a menor según la varianza
    orden = sortperm(vals, rev=true)
    vals = vals[orden]
    vecs = vecs[:, orden]
    
    # 5. seleccionar k componentes principales
    k = min(k, size(vecs, 2))
    P = vecs[:, 1:k]
    
    # 6. proyectar los datos sobre los componentes principales
    proyeccion = A_centrada * P
    
    # 7. reconstruir la matriz aproximada
    A_rec = proyeccion * P' .+ mean(A, dims=1)
    
    return A_rec
end
```

> Reconstrucción de la imagen.

```julia
function comprimir_imagen_pca_cov(ruta_imagen::String)
    println("Cargando imagen desde: ", ruta_imagen)
    try
        img = load(ruta_imagen)
        println("Imagen cargada con tamaño ", size(img))

        R = Float32.(red.(img))
        G = Float32.(green.(img))
        B = Float32.(blue.(img))

        base_dim = min(size(R,1), size(R,2))
        max_k = min(base_dim, 512)
        ks = collect(potencias_de_dos(max_k))
        if isempty(ks)
            ks = [1]
        end
        println("Componentes a probar: ", ks)

        p_original = plot(img, title="Original", aspect_ratio=:equal, axis=nothing)
        display(p_original)
        sleep(1)

        plots_list = [p_original]

        for k in ks
            kk = min(k, base_dim)

            r_rec = pca_cov_imagen(R, kk)
            g_rec = pca_cov_imagen(G, kk)
            b_rec = pca_cov_imagen(B, kk)

            img_rec = colorview(RGB,
                                clamp.(r_rec, 0f0, 1f0),
                                clamp.(g_rec, 0f0, 1f0),
                                clamp.(b_rec, 0f0, 1f0))

            p = plot(img_rec, title="k = $kk", aspect_ratio=:equal, axis=nothing)
            display(p)
            push!(plots_list, p)

            sleep(1)
        end

        n_plots = length(plots_list)
        n_cols = ceil(Int, sqrt(n_plots))
        n_rows = ceil(Int, n_plots / n_cols)

        display(plot(plots_list..., layout=(n_rows, n_cols), size=(1400, 900)))

    catch e
        if e isa SystemError
            println("\nNo se pudo encontrar la imagen en la ruta especificada.")
        else
            rethrow(e)
        end
    end
end
```

- Correlación preserva mejor la proporción y el balance entre los colores al normalizar las variables.

- Covarianza resalta las zonas de mayor variabilidad o contraste, lo que puede mejorar el detalle en regiones con cambios intensos de color o luminosidad.

Algo importante es conocer la aplicación que le daremos a los modelos, esto puede sugerir que necesitemos más componentes en algún contexto. Es decir, existen aplicaciones de alta precisión (como reconocimiento facial o análisis médico de imágenes) donde sí conviene usar más componentes.

Por útlimo, no se trata solo de reducir dimensiones, sino de entender qué información es relevante para la tarea específica que hacemos. No existe un número fijo de componentes óptimo. El valor del PCA radica precisamente en esa flexibilidad para adaptarse al propósito del estudio.
